# Step 6 : Data Visualization

```python
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category',data=samplel)
plt.title('Catrgory vs Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Category')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15,15))
sns.countplot(x=samplel['State'])
plt.xticks(rotation=90)
plt.title("STATE")
plt.show()
```

```python
Profit_plot = (ggplot(sample, aes(x='Sub-Category', y='Profit', fill ='Sub-Category')) + geom_col() + coord_flip() 
               + scale_fill_brewer(type= 'div', palette= "Spectral") + theme_classic() + ggtitle('Pie chart'))

display(Profit_plot)
```

The above pie chart shows the profit and loss of each and every Subcategories.

```python
sns.set(style="whitegrid")
plt.figure(2, figsize=(20,15))
sns.barplot(x='Sub-Category',y='Profit', data=sample, palette='Spectral')
plt.suptitle('Pie Consumption Patterns in the United States', fontsize=16)
plt.show()
```

```python
ggplot(sample, aes(x='Ship Mode', fill='Category')) + geom_bar(stat='count')
```

```python
figsize=(15,10)
sns.pairplot(samplel,hue='Sub-Category')
plt.show()
```

```python
x = sample.groupby(by="Category")[['Profit']].sum()
b=x.sort_values(by=['Profit'],ascending=True)
b.plot(kind='bar')
plt.show()
```

We can see that the above graph that the profit earned from furniture is the lowest, the profit earned from techanology is the highest, and the office supplies profit is somewhat similar to techanology, now lets focus 

```python
x = sample.groupby(by="Sub-Category")[['Discount']].sum()
b=x.sort_values(by=['Discount'],ascending=True)
b.plot(kind='bar')

x = sample.groupby(by="Sub-Category")[['Profit']].sum()
b=x.sort_values(by=['Profit'],ascending=True)
b.plot(kind='bar')

plt.show()
```

```python
fig = plt.figure(figsize=(12,6))
a = fig.add_subplot(1,2,1)
sns.lineplot(x = "Discount", y = "Profit",ax=a, data = sample, color = "red")
#plt.show()

#fig = plt.figure(figsize=(7,5))
b = fig.add_subplot(1,2,2)
sns.lineplot(x = "Sales", y = "Profit",ax=b, data = sample, color = "green")
plt.tight_layout(pad = 8.0)
plt.show()
```

Here we can say that profit is decreasing with increament in Discount.
Also, profit is not very much depending the nubmer of sales, there are other factors which have dominating effect on profit

## Summary
